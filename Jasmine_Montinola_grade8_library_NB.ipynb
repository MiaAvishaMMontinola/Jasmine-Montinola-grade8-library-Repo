```python
# Install requests if needed
import requests
import json

# Replace with your actual raw GitHub URL
url =  "https://raw.githubusercontent.com/MiaAvishaMMontinola/Jasmine-Montinola-grade8-library-Repo/refs/heads/main/books.json"

response = requests.get(url)
books_data = response.json()

print("Books JSON loaded successfully!")
print(books_data)
```

```python
import firebase_admin
from firebase_admin import credentials, db

cred = credentials.Certificate("firebase_key.json")

# Initialize app only if it hasn't been initialized already
if not firebase_admin._apps:
    firebase_admin.initialize_app(cred, {
        'databaseURL': 'https://montinola-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/'
    })
    print("Firebase connected successfully!")
else:
    print("Firebase app already initialized.")
```

```python
ref = db.reference('books')
ref.set(books_data)

print("Books uploaded to Firebase Realtime Database!")
```

```python
# Inspect the content of firebase_key.json
print("--- Content of firebase_key.json ---")
with open('firebase_key.json', 'r') as f:
    print(f.read())
print("------------------------------------")
```

```python
books_from_db = ref.get()
print("All Books Data:")
print(books_from_db)
```

```python
print("Book Titles:")
for book in books_from_db['books']:
    print(book['title'])
```

```python
print("Title and Availability:")
for book in books_from_db['books']:
    print(book['title'], "-", book['details']['available'])
```

```python
total_books = len(books_from_db['books'])
print("Total number of books in the library:", total_books)
```
